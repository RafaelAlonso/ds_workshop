```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
nyc_df = pd.read_csv('nyc_airbnb/AB_NYC_2019.csv')
```

```python
nyc_df.columns
```

```python
nyc_df.head()
```

```python
columns_to_drop = ['id', 'host_name', 'last_review']
nyc_df.drop(columns_to_drop, axis="columns", inplace=True)
```

```python
nyc_df.fillna({'reviews_per_month': 0}, inplace=True)
```

```python
nyc_df['neighbourhood'].value_counts().head().plot(kind="bar")
```

```python
sns.countplot(data=nyc_df, x='neighbourhood')
```

```python
sns.countplot(data=nyc_df, x='neighbourhood_group')
```

```python
nyc_df[['neighbourhood_group', 'room_type']].value_counts()
```

```python
nyc_df[['neighbourhood_group', 'room_type']].value_counts().plot(kind='bar')
```

```python
nyc_df.nsmallest(10, 'reviews_per_month')
```

```python
sns.violinplot(data=nyc_df[nyc_df['price'] < 500], y="neighbourhood_group", x="price")
```

```python
nyc_df.plot(
    kind='scatter',
    x='longitude',
    y='latitude',
    c='price',
    cmap='inferno',
    colorbar=True,
    alpha=0.8,
    figsize=(12,8))
```
